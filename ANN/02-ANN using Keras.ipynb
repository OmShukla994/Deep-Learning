```python
X = dataset.iloc[:, 3:13].values # X = from CreditScore to EstimatedSalary
y = dataset.iloc[:, 13].values # Y = Exited
```

```python
#OneHotEncoder is same like dummy.
#OneHotEncoder is useful than Dummy Creation , when we have large set of data as it converts it into matrix 
#rather than dataframe and matrix is faster.
```

```python
#LabelEncoder is used for variable having only 2 classes like Gender
#OneHotEncoder is when there are more than 2 classes.But always LabelEncoder is used first and then convert to OneHotEncoder.EX:-

from sklearn.preprocessing import LabelEncoder, OneHotEncoder


labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1]) #converting 'Geography' to 0,1,2 for ['France', 'Spain', 'Germany']
                                                  #using label encoder.

labelencoder_X_2 = LabelEncoder()   
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2]) #converting 'Gender' to 0, 1 for ['Male','Female']
```

```python
X
```

```python
#Need ColumnTransformer function to specify column indexes for encoding.Otherwise, onehotencoder transforms all interger colums

from sklearn.compose import ColumnTransformer

onehotencoder = ColumnTransformer(
  [('one_hot_encoder', OneHotEncoder(categories='auto'), [1])],   # The column numbers to be transformed (here is [1] but can be any index)
 remainder='passthrough')                                         # Leave the rest of the columns untouched


X = onehotencoder.fit_transform(X)

X = X[:, 1:]

X_df=pd.DataFrame(X)
X_df.head(10)
```

```python
#keras is intercompatible with sklearn.KERAS has extra step i.e., compilation

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### Modelling By using Keras liberary  

```python
# Importing the Keras libraries and packages
import keras #it is from tensorflow so. you might install tensorflow and then keras.
from keras.models import Sequential #Sequential,each layer passes info to its next layer.

from keras.layers import Dense #all nodes connected with all nodes in next layer.
```

```python
#Initializing Neural Network
classifier=Sequential()
```

```python
#ARCHITECTURE OF THE MODEL
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'softmax', input_dim = 11))
# output_dim = 8 nodes in all hidden layers bcoz of the formula 
#and input_dim = 11 bcoz its equal to no. of independent vars.

# Adding the second hidden layer
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'softmax')) #change output_dim and activation and check
                                                                            #init = uniform is intialising weights as uniform.
# Adding the third layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'softmax'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling Neural Network
#optimizer tunes the learning rate automatically
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', weighted_metrics = ['accuracy'])
```

```python
# Fitting our model 
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
#(8000/100) * 100 = 8000 times weight is updated
```

```python
# Predicting the Test set results (it is unbalanced data,you shouls balance it first and then find cutoff)
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.2)   #---using response rate[405/2000 ~ 0.2 (for 1's in data)] for quick estimation of optimal cut-off. 
```

```python
# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import sklearn.metrics as metrics
print(metrics.classification_report(y_test,y_pred))
```

```python
metrics.roc_auc_score(y_test, y_pred) #auc cutoff is 80%
```
